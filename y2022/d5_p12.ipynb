```python
from copy import deepcopy
import re
import textwrap
```

```python
import io
test_in = """    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2"""
IN = io.StringIO(test_in)
IN = open("d5_in.txt")

with IN as f:
    stacks_txt, moves = f.read().split("\n\n")
    stacks_txt = stacks_txt.split("\n")
    moves = [m[5:] for m in moves.split("\n")]
patt_m = re.compile(r"[a-z ]+")
MOVES = [
    [int(x) for x in re.split(patt_m, m)] for m in moves
]
num_stacks = len(re.split(r" +", stacks_txt[-1].strip(" ")))
stacks = [[] for _ in range(num_stacks)]
for line in stacks_txt[:-1]:
    elements = textwrap.wrap(line, 4, drop_whitespace=False)
    for idx, e in enumerate(elements):
        if (e_ := e.strip(" []")):
            stacks[idx].append(e_)
STACKS = [s[::-1] for s in stacks]
```

Part 1

```python
stacks = deepcopy(STACKS)
for m in MOVES:
    num, from_, to_ = m[0], m[1]-1, m[2]-1
    for i in range(num):
        x = stacks[from_].pop()
        stacks[to_].append(x)
"".join([s[-1] for s in stacks])
```

Part 2

```python
stacks = deepcopy(STACKS)
for m in MOVES:
    num, from_, to_ = m[0], m[1]-1, m[2]-1
    tmp = []
    for i in range(num):
        x = stacks[from_].pop()
        tmp.append(x)
    tmp = tmp[::-1]
    stacks[to_] += tmp
"".join([s[-1] for s in stacks])
```
