Part 1

```python
def is_in_right_order(left: list, right: list) -> bool:
    if not left and right:
        return True
    elif left and not right:
        return False
    elif not left and not right:
        return False
    assert left and right
    l1, r1 = left[0], right[0]
    if isinstance(l1, int) and isinstance(r1, int):
        if l1 < r1:
            return True
        elif l1 > r1:
            return False
        else:
            return is_in_right_order(left[1:], right[1:])
    elif isinstance(l1, list) and isinstance(r1, list):
        if l1 == r1:
            return is_in_right_order(left[1:], right[1:])
        else:
            return is_in_right_order(l1, r1)
    elif isinstance(l1, list) and isinstance(r1, int):
        return is_in_right_order([l1] + left[1:], [[r1]] + right[1:])
    elif isinstance(r1, list) and isinstance(l1, int):
        return is_in_right_order([[l1]] + left[1:], [r1] + right[1:])
    raise ValueError("Shouldn't happen")
```

```python
pairs = deepcopy(PAIRS)

total = 0
for i, pair in enumerate(pairs, start=1):
    left, right = pair
    if is_in_right_order(left, right):
        total += i
total
```

```python
pairs = deepcopy(PAIRS)

total_2, total_6 = 0, 0
for i, pair in enumerate(pairs, start=1):
    p1, p2 = pair
    if is_in_right_order(p1, [[2]]):
        total_2 += 1
    if is_in_right_order(p2, [[2]]):
        total_2 += 1
    if is_in_right_order(p1, [[6]]):
        total_6 += 1
    if is_in_right_order(p2, [[6]]):
        total_6 += 1
(total_2 + 1)*(total_6+2)
```
