```python
# import io
# test_in = """R 4
# U 4
# L 3
# D 1
# R 4
# D 1
# L 5
# R 2"""
# IN = io.StringIO(test_in)
# test_in_2 = """R 5
# U 8
# L 8
# D 3
# R 17
# D 10
# L 25
# U 20"""
# IN = io.StringIO(test_in_2)
IN = open("d9_in.txt")

with IN as f:
    DIRS = [x.split(" ") for x in f.read().split("\n")]
```

Part 1

```python
def update(dir, old_h, old_t):
    hx, hy = old_h
    tx, ty = old_t
    match dir:
        case 'U':
            new_h = hx, hy+1
            if ty == hy-1:
                new_t = (hx, hy)
            else:
                new_t = old_t
        case 'D':
            new_h = hx, hy-1
            if ty == hy+1:
                new_t = (hx, hy)
            else:
                new_t = old_t
        case 'R':
            new_h = hx+1, hy
            if tx == hx-1:
                new_t = (hx, hy)
            else:
                new_t = old_t
        case 'L':
            new_h = hx-1, hy
            if tx == hx+1:
                new_t = (hx, hy)
            else:
                new_t = old_t
    return new_h, new_t
```

```python
h, t = (0, 0), (0, 0)
tail_pos = {t}
for d, n in DIRS:
    n = int(n)
    for k in range(n):
        h, t = update(d, h, t)
        tail_pos |= {t}
len(tail_pos)
```

Part 2

```python
def update_head(dir, old_pos):
    x, y = old_pos
    match dir:
        case "U":
            return (x, y+1)
        case "D":
            return (x, y-1)
        case "L":
            return (x-1, y)
        case "R":
            return (x+1, y)

def update_trail(new_front, old_back):
    xf, yf = new_front
    xb, yb = old_back
    dx = xf - xb
    dy = yf - yb
    assert abs(dx)+abs(dy) <= 4, f"new_front: {xf}, {yf}, old_back: {xb}, {yb}"
    if (dx, dy) == (0, 2):
        return (xb, yb+1)
    elif (dx, dy) == (0, -2):
        return (xb, yb-1)
    elif (dx, dy) == (2, 0):
        return (xb+1, yb)
    elif (dx, dy) == (-2, 0):
        return (xb-1, yb)
    elif (dx, dy) in [(2, 1), (1, 2), (2, 2)]:
        return (xb+1, yb+1)
    elif (dx, dy) in [(2, -1), (1, -2), (2, -2)]:
        return (xb+1, yb-1)
    elif (dx, dy) in [(-2, -1), (-1, -2), (-2, -2)]:
        return (xb-1, yb-1)
    elif (dx, dy) in [(-2, 1), (-1, 2), (-2, 2)]:
        return (xb-1, yb+1)
    return (xb, yb)
```

```python
h, t1, t2, t3, t4, t5, t6, t7, t8, t9 = 10*[(0, 0)]
tail_pos = {t9}
for d, n in DIRS:
    n = int(n)
    for k in range(n):
        h = update_head(d, h)
        t1 = update_trail(new_front=h, old_back=t1)
        t2 = update_trail(new_front=t1, old_back=t2)
        t3 = update_trail(new_front=t2, old_back=t3)
        t4 = update_trail(new_front=t3, old_back=t4)
        t5 = update_trail(new_front=t4, old_back=t5)
        t6 = update_trail(new_front=t5, old_back=t6)
        t7 = update_trail(new_front=t6, old_back=t7)
        t8 = update_trail(new_front=t7, old_back=t8)
        t9 = update_trail(new_front=t8, old_back=t9)
        tail_pos |= {t9}
len(tail_pos)
```
