```python
# import io
# test_in = """2-4,6-8
# 2-3,4-5
# 5-7,7-9
# 2-8,3-7
# 6-6,4-6
# 2-6,4-8"""
# IN = io.StringIO(test_in)
IN = open("d4_in.txt")

with IN as f:
    lines = [line.split(",") for line in f.read().split("\n")]
    ranges = [
        [[int(x) for x in r.split("-")] for r in line]
        for line in lines
    ]
```

Part 1

```python
def contains(left, right):
    """True if left contains right, left <= right lexicographically"""
    # left      xl         yl       OR       xl     yl
    # right        xr    yr                  xr        yr
    xl, yl = left
    xr, yr = right
    return (
        (xl < xr and yl >= yr)
        or (xl == xr and yl <= yr)
    ) 
```

```python
total = 0
for range_pair in ranges:
    left, right = sorted(range_pair)
    total += contains(left, right)
total
```

Part 2

```python
def overlaps(left, right):
    """True if left overlaps with right, left <= right lexicographically"""
    # left      xl         yl       OR       xl     yl
    # right        xr    yr                  xr        yr
    xl, yl = left
    xr, yr = right
    return xl <= xr <= yl
```

```python
total = 0
for range_pair in ranges:
    left, right = sorted(range_pair)
    total += overlaps(left, right)
total
```
