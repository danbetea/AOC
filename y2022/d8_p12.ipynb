Part 2

```python
PATT = re.compile(r"^1+")
def dist(arr, patt=PATT):
    s = "".join((arr > 0).astype(np.int32).astype(str).tolist())
    if (m := re.match(patt, s)) is not None:
        return 1 + m.end()
    return 1

def score(mat, i, j, rows, cols, patt=PATT):
    el = mat[i, j]
    top = mat[:i, j][::-1]
    bottom = mat[i+1:, j]
    left = mat[i, :j][::-1]
    right = mat[i, j+1:]
    d1 = min(dist(el - top, patt), i)
    d2 = min(dist(el - bottom, patt), (rows-1)-(i+1)+1)
    d3 = min(dist(el - left, patt), j)
    d4 = min(dist(el - right, patt), (cols-1)-(j+1)+1)
    return d1 * d2 * d3 * d4
```

```python
rows, cols = MAT_NP.shape
max_ = 0
for i in range(1, rows-1):
    for j in range(1, cols-1):
        m = score(MAT_NP, i, j, rows, cols)
        if m > max_:
            max_ = m
max_
```
