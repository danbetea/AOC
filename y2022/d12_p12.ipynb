```python
from queue import PriorityQueue
```

```python
# import io
# test_in = """Sabqponm
# abcryxxl
# accszExk
# acctuvwj
# abdefghi"""
# IN = io.StringIO(test_in)
IN = open("d12_in.txt")

with IN as f:
    MAP = [list(m) for m in f.read().split("\n")]
```

```python
def is_in_bounds(u, rows, cols):
    i, j = u
    return 0 <= i < rows and 0 <= j < cols


def new_ord(symb):
    if symb == "S":
        return ord('a')
    elif symb == "E":
        return ord('z')
    return ord(symb) 


def neighbors(u, map_, rows, cols):
    i, j = u
    all_neighbors = [(i, j-1), (i, j+1), (i+1, j), (i-1, j)]
    return [
        n for n in all_neighbors
        if is_in_bounds(n, rows, cols)
        and new_ord(map_[n[0]][n[1]]) - new_ord(map_[i][j]) <= 1
    ]


def find_point(map_, symb):
    for i in range(len(map_)):
        for j in range(len(map_[0])):
            if map_[i][j] == symb:
                return i, j
    raise ValueError("Source not found")
```

Part 1

```python
def shortest_path(map_, source):
    rows, cols = len(map_), len(map_[0])
    dist = {(i, j): 1_000_000 for i in range(rows) for j in range(cols)}
    prev = {(i, j): None for i in range(rows) for j in range(cols)}
    Q = PriorityQueue()
    dist[source] = 0
    Q.put((dist[source], source))

    while not Q.empty():
        d_u, u = Q.get()
        for v in neighbors(u, map_, rows, cols):
            alt = d_u + 1
            if alt < dist[v]:
                dist[v] = alt
                prev[v] = u
                Q.put((dist[v], v))
    return dist, prev
```

```python
map_ = MAP
source = find_point(map_, symb="S")
dest = find_point(map_, symb='E')
dist, prev = shortest_path(map_, source)
dist[dest]
```

```python
map_ = MAP
sources = []
for i in range(len(map_)):
    for j in range(len(map_[0])):
        if map_[i][j] in ['S', 'a']:
            sources.append((i, j))
```

```python
min_dists = []
dest = find_point(map_, symb='E')
for i, s in enumerate(sources):
    dist, prev = shortest_path(map_, s)
    min_dists.append(dist[dest])
min(min_dists)
```
