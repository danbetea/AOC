```python
# import io
# test_in = """A Y
# B X
# C Z"""
# IN = io.StringIO(test_in)
IN = open("d2_in.txt")

with IN as f:
    lines = [
        line.split(" ")
        for line in f.read().split("\n")
    ]
```

Part 1

```python
def score_p1(pair: list[str]):
    match pair:
        case "A", "X":
            return 3 + 1
        case "A", "Y":
            return 6 + 2
        case "A", "Z":
            return 0 + 3
        case "B", "X":
            return 0 + 1
        case "B", "Y":
            return 3 + 2
        case "B", "Z":
            return 6 + 3
        case "C", "X":
            return 6 + 1
        case "C", "Y":
            return 0 + 2
        case "C", "Z":
            return 3 + 3
```

```python
sum(score_p1(line) for line in lines)
```

Part 2

```python
def score_p2(pair: list[str]):
    match pair:
        case "A", "X":
            return 0 + 3
        case "A", "Y":
            return 3 + 1
        case "A", "Z":
            return 6 + 2
        case "B", "X":
            return 0 + 1
        case "B", "Y":
            return 3 + 2
        case "B", "Z":
            return 6 + 3
        case "C", "X":
            return 0 + 2
        case "C", "Y":
            return 3 + 3
        case "C", "Z":
            return 6 + 1
```

```python
sum(score_p2(line) for line in lines)
```
