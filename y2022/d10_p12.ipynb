Part 1

```python
special_cycles = [20, 60, 100, 140, 180, 220][::-1]
cycle = 1
X = 1
ops = deepcopy(OPS)

total = 0
while cycle <= 1_000_000:

    two_op = False
    if not special_cycles:
        break

    op = ops.pop()
    if op == "noop":
        val = 0
        cycle += 1
    else:
        val = int(op.split(" ")[1])
        two_op = True
        cycle += 2
    
    old_X = X
    new_X = X + val

    if ((cycle == special_cycles[-1] and not two_op)
        or (cycle == special_cycles[-1] and two_op)):
        c = special_cycles.pop()
        total += new_X * c
    elif (cycle-1 == special_cycles[-1] and two_op):
        c = special_cycles.pop()
        total += old_X * c
    
    X = new_X
total
```

Part 2

```python
special_cycles = list(range(2, 241))[::-1]
cycle = 1
X = 1
ops = deepcopy(OPS)
Xs = []
new_ops = []
for o in ops:
    if o == "noop":
        new_ops.append("addx 0")
    else:
        new_ops.append(o)
        new_ops.append("addx 0")
new_ops

cycle = 1
while cycle <= 240:
    Xs.append((cycle, X))
    o = new_ops.pop()
    X += int(o.split(" ")[1])
    cycle += 1
```

```python
screen = [["." for x in range(40)] for y in range(6)]
for i, X in Xs:
    vals = [X-1, X, X+1]
    if 1 <= i <= 40 and (i-1) in vals:
        screen[0][i-1] = "#"
    elif 41 <= i <= 80 and (i-41) in vals:
        screen[1][i-41] = "#"
    elif 81 <= i <= 120 and (i-81) in vals:
        screen[2][i-81] = "#"
    elif 121 <= i <= 160 and (i-121) in vals:
        screen[3][i-121] = "#"
    elif 161 <= i <= 200 and (i-161) in vals:
        screen[4][i-161] = "#"
    elif 201 <= i <= 240 and (i-201) in vals:
        screen[5][i-201] = "#"
print("\n".join(["".join(s) for s in screen]))
```
