```python
import io
test_in = """vJrwpWtwJgWrhcsFMMfFFhFp
jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL
PmmdzqPrVvPwwTWBwg
wMqvLMZHhHMvwLHjbvcjnnSBnvTQFn
ttgJtRGJQctTZtZT
CrZsJsPPZsGzwwsLwLmpwMDw"""
IN = io.StringIO(test_in)
IN = open("d3_in.txt")

with IN as f:
    lines = f.read().split("\n")
    lines_p1 = [
        [line[:len(line)//2], line[len(line)//2:]]
    for line in lines]
    lines_p2 = [
        [lines[3*i], lines[3*i+1], lines[3*i+2]]
        for i in range(len(lines) // 3)
    ]
```

Part 1

```python
def priority(a: str) -> int:
    return ord(a) - 96 if a.islower() else ord(a) - 38
```

```python
total = 0
for pair in lines_p1:
    s1, s2 = set(pair[0]), set(pair[1])
    common = s1 & s2
    assert len(common) == 1
    item, = common
    total += priority(item)
total
```

Part 2

```python
total = 0
for triple in lines_p2:
    s1, s2, s3 = set(triple[0]), set(triple[1]), set(triple[2])
    common = s1 & s2 & s3
    assert len(common) == 1
    item, = common
    total += priority(item)
total
```
