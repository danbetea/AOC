```python
from collections import deque
from copy import deepcopy
```

```python
MONKEYS = {
    0: {
        'items': deque([89, 95, 92, 64, 87, 68]),
        'op': lambda x: x*11,
        'test': lambda x: x % 2 == 0,
        'true_m': 7,
        'false_m': 4,
    },
    1: {
        'items': deque([87, 67]),
        'op': lambda x: x+1,
        'test': lambda x: x % 13 == 0,
        'true_m': 3,
        'false_m': 6,
    },
    2: {
        'items': deque([95, 79, 92, 82, 60]),
        'op': lambda x: x+6,
        'test': lambda x: x % 3 == 0,
        'true_m': 1,
        'false_m': 6,
    },
    3: {
        'items': deque([67, 97, 56]),
        'op': lambda x: x*x,
        'test': lambda x: x % 17 == 0,
        'true_m': 7,
        'false_m': 0,
    },
    4: {
        'items': deque([80, 68, 87, 94, 61, 59, 50, 68]),
        'op': lambda x: x*7,
        'test': lambda x: x % 19 == 0,
        'true_m': 5,
        'false_m': 2,
    },
    5: {
        'items': deque([73, 51, 76, 59]),
        'op': lambda x: x+8,
        'test': lambda x: x % 7 == 0,
        'true_m': 2,
        'false_m': 1,
    },
    6: {
        'items': deque([92]),
        'op': lambda x: x+5,
        'test': lambda x: x % 11 == 0,
        'true_m': 3,
        'false_m': 0,
    },
    7: {
        'items': deque([99, 76, 78, 76, 79, 90, 89]),
        'op': lambda x: x+7,
        'test': lambda x: x % 5 == 0,
        'true_m': 4,
        'false_m': 5,
    },
}

# MONKEYS = {
#     0: {
#         'items': deque([79, 98]),
#         'op': lambda x: x*19,
#         'test': lambda x: x % 23 == 0,
#         'true_m': 2,
#         'false_m': 3,
#     },
#     1: {
#         'items': deque([54, 65, 75, 74]),
#         'op': lambda x: x+6,
#         'test': lambda x: x % 19 == 0,
#         'true_m': 2,
#         'false_m': 0,
#     },
#     2: {
#         'items': deque([79, 60, 97]),
#         'op': lambda x: x*x,
#         'test': lambda x: x % 13 == 0,
#         'true_m': 1,
#         'false_m': 3,
#     },
#     3: {
#         'items': deque([74]),
#         'op': lambda x: x+3,
#         'test': lambda x: x % 17 == 0,
#         'true_m': 0,
#         'false_m': 1,
#     },
# }
```

Part 1

```python
round = 0
num_rounds = 20
monkeys = deepcopy(MONKEYS)
counters = [0 for _ in range(len(monkeys))]

while round < num_rounds:
    for i in range(len(monkeys)):
        monkey = monkeys[i]
        queue = monkey['items']
        while queue:
            item = queue.popleft()
            item = monkey['op'](item)
            item //= 3
            new_monkey = (
                monkeys[monkey['true_m']] if monkey['test'](item)
                else monkeys[monkey['false_m']]
            )
            new_monkey['items'].append(item)
            counters[i] += 1
    round += 1
```

```python
x, y = sorted(counters, reverse=True)[:2]
x * y
```

Part 2

```python
round = 0
num_rounds = 10_000
monkeys = deepcopy(MONKEYS)
counters = [0 for _ in range(len(monkeys))]
# product for Chinese Remainder Theorem
N = 2*13*3*17*19*7*11*5


while round < num_rounds:
    for i in range(len(monkeys)):
        monkey = monkeys[i]
        queue = monkey['items']
        while queue:
            item = queue.popleft()
            item = monkey['op'](item)
            item %= N
            new_monkey = (
                monkeys[monkey['true_m']] if monkey['test'](item)
                else monkeys[monkey['false_m']]
            )
            new_monkey['items'].append(item)
            counters[i] += 1
    round += 1
```

```python
x, y = sorted(counters, reverse=True)[:2]
x * y
```
