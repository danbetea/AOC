```python
from functools import cache
```

```python
# from io import StringIO

# for part 1
# IN_ = """aaa: you hhh
# you: bbb ccc
# bbb: ddd eee
# ccc: ddd eee fff
# ddd: ggg
# eee: out
# fff: out
# ggg: out
# hhh: ccc fff iii
# iii: out"""
# IN = StringIO(IN_)

# for part 2
# IN = """svr: aaa bbb
# aaa: fft
# fft: ccc
# bbb: tty
# tty: ccc
# ccc: ddd eee
# ddd: hub
# hub: fff
# eee: dac
# dac: fff
# fff: ggg hhh
# ggg: out
# hhh: out"""
# IN = StringIO(IN)

IN = open("d11_in.txt")

with IN as f:
    DAG = {}
    lines = f.read().split("\n")
    for line in lines:
        source, _, dests = line.partition(": ")
        DAG[source] = tuple(dests.split(" "))
```

Part 1

```python
@cache
def count(source, dest):
    res = 0
    for v in DAG[source]:
        if v == dest:
            return 1
        else:
            res += count(v, dest)
    return res
```

```python
count("you", "out")
```

Part 2

```python
@cache
def count_2(source, dest, seen_fft, seen_dac):
    res = 0
    for v in DAG[source]:
        if v == dest and seen_fft and seen_dac:
            return 1
        elif v == dest and not (seen_fft and seen_dac):
            return 0
        elif v == "fft":
            res += count(v, dest, True, seen_dac)
        elif v == "dac":
            res += count(v, dest, seen_fft, True)
        else:
            res += count(v, dest, seen_fft, seen_dac)
    return res
```

```python
count_2("svr", "out", False, False)
```
