```python
# IN = io.StringIO(test_in)
IN = open("d1_in.txt")

clicks = []
start = 50
with IN as f:
    lines = f.readlines()
    lines = [line.rstrip("\r\n") for line in lines]
for line in lines:
    d, n = line[0], int(line[1:])
    clicks.append((d, n))
```

Part 1

```python
start = 50
count = 0
for d, n in clicks:
    if d == "R":
        start = (start + n) % 100
    else:
        start = (start - n) % 100
    if start == 0:
        count += 1
count
```

Part 2

```python
start = 50
count = 0
for d, n in clicks:
    old_n = n
    if d == "R":
        sum_ = start + n
        start = sum_ % 100
        count += sum_ // 100
    else:
        diff = start - n
        # diff == start == 0 is impossible, otherwise this won't work
        count += (-diff // 100 + 1) - (1 if start == 0 else 0)
        start = diff % 100
count
```
