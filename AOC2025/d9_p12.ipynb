```python
from itertools import combinations
```

```python
from io import StringIO
IN_ = """7,1
11,1
11,7
9,7
9,5
2,5
2,3
7,3"""
IN = StringIO(IN_)
IN = open("d9_in.txt")
with IN as f:
    lines = f.read().split("\n")
    POINTS = [line.split(",") for line in lines]
    POINTS = [tuple([int(x) for x in line]) for line in POINTS]
```

```python
def area(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    min_x, max_x = min(x1, x2), max(x1, x2)
    min_y, max_y = min(y1, y2), max(y1, y2)
    return (max_x - min_x + 1) * (max_y - min_y + 1)

def area_shifted(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    min_x, max_x = min(x1, x2), max(x1, x2)
    min_y, max_y = min(y1, y2), max(y1, y2)
    return int(max_x - min_x) * int(max_y - min_y)

def get_all_points(points):
    out = []
    for i in range(len(points)):
        p = points[i]
        q = points[0] if i == len(points) - 1 else points[i+1] 
        out.append(p)
        px, py = p
        qx, qy = q
        if px == qx:
            if py < qy:
                for i in range(py+1, qy):
                    out.append([px, i])
            else:  # qy < py
                for i in range(py-1, qy, -1):
                    out.append([px, i])
        else:  # py == qy
            if px < qx:
                for i in range(px+1, qx):
                    out.append([i, py])
            else:  # px > qx
                for i in range(px-1, qx, -1):
                    out.append([i, py])
    return out
```

Part 1

```python
max(area(pair[0], pair[1]) for pair in combinations(POINTS, 2))
```

```python
bad_pairs = set()
areas = []
for pair in combinations(POINTS, 2):
    is_good = True
    p, q = pair
    px, py = p
    qx, qy = q
    min_x, max_x = min(px, qx), max(px, qx)
    min_y, max_y = min(py, qy), max(py, qy)
    for m in POINTS:
        if m == p or m == q:
            continue
        mx, my = m
        if (
            (min_x < mx < max_x and min_y <= my <= max_y)
            or (min_x <= mx <= max_x and min_y < my < max_y)   
        ):
            bad_pairs |= {pair}
            is_good = False
            break
    if is_good:
        areas.append(area(p, q))
```

```python
max(areas)
```
