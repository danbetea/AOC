```python
from itertools import combinations
```

```python
# from io import StringIO
# IN_ = """7,1
# 11,1
# 11,7
# 9,7
# 9,5
# 2,5
# 2,3
# 7,3"""
# IN = StringIO(IN_)
IN = open("d9_in.txt")
with IN as f:
    lines = f.read().split("\n")
    POINTS = [line.split(",") for line in lines]
    POINTS = [tuple([int(x) for x in line]) for line in POINTS]
```

```python
def area(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    min_x, max_x = min(x1, x2), max(x1, x2)
    min_y, max_y = min(y1, y2), max(y1, y2)
    return (max_x - min_x + 1) * (max_y - min_y + 1)

def compute_segments(points):
    verticals, horizontals = {p: set() for p in points}, {p: set() for p in points}
    points_ = points + [points[0]]
    for i in range(len(points_)-1):
        p = points_[i]
        q = points_[i+1]
        px, py = p
        qx, qy = q
        if px == qx:
            verticals[p] |= {(min(py, qy), max(py, qy))}
            verticals[q] |= {(min(py, qy), max(py, qy))}
        else:  # py == qy
            horizontals[p] |= {(min(px, qx), max(px, qx))}
            horizontals[q] |= {(min(px, qx), max(px, qx))}
    return horizontals, verticals

def is_empty_strict_intersection(i, j):
    i1, i2 = i
    j1, j2 = j
    assert i1 <= i2
    assert j1 <= j2
    return i2 <= j1 or j2 <= i1
```

Part 1

```python
max(area(pair[0], pair[1]) for pair in combinations(POINTS, 2))
```

Part 2

```python
#              1111
#    01234567890123
#  0 ..............
#  1 .......#XXX#..
#  2 .......X...X..
#  3 ..#XXXX#...X..
#  4 ..X........X..
#  5 ..#XXXXXX#.X..
#  6 .........X.X..
#  7 .........#X#..
#  8 ..............

# test of a complicated polygon
# the two Xs cannot be the corners of a rectangle
#                       
#                     +------+ 
#                     |      |
#  +------------------+      X------------+
#  |                                      |
#  |   +----------------------------------+
#  |   |
#  |   +------------------------------+
#  |                                  |
#  +-------X    +---------------------+
#          |    |
#          +----+
```

```python
horizontals, verticals = compute_segments(POINTS)

bad_pairs = set()
areas = []
for pair in combinations(POINTS, 2):
    is_good = True
    p, q = pair
    px, py = p
    qx, qy = q
    min_x, max_x = min(px, qx), max(px, qx)
    min_y, max_y = min(py, qy), max(py, qy)
    for m in POINTS:
        if m == p or m == q:
            continue
        mx, my = m
        if (min_x < mx < max_x and min_y < my < max_y):
            bad_pairs |= {pair}
            is_good = False
            break
        elif (min_y < my < max_y):
            for l in horizontals[m]:
                if not is_empty_strict_intersection(l, (min_x, max_x)):
                    bad_pairs |= {pair}
                    is_good = False
                    break
            if not is_good:
                break
        elif (min_x < mx < max_x):
            for l in verticals[m]:
                if not is_empty_strict_intersection(l, (min_y, max_y)):
                    bad_pairs |= {pair}
                    is_good = False
                    break
            if not is_good:
                break
    if is_good:
        areas.append(area(p, q))
```

```python
max(areas)
```
