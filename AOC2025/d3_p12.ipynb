```python
val = 0
for l in lines:
    m = max(l[:-1])
    nums = []
    for j in range(len(l)-1):
        if l[j] == m:
            n = max(l[j+1:])
            break
    num = m*10+n
    val += num
val
```

```python
lst = []
lst_tups = [[] for i in range(len(lines))]
for idx, l in enumerate(lines):
    d = {i: [] for i in range(9, 0, -1)}
    for i, k in enumerate(l):
        d[k].append(i)
        lst_tups[idx].append((k, i))
    lst.append(d)
    for i in d:
        d[i] = sorted(d[i], reverse=True)
```
