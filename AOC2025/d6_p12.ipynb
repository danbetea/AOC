```python
from functools import reduce
import re


import numpy as np
```

Part 1

```python
# from io import StringIO
# IN_ = """123 328  51 64 
#  45 64  387 23 
#   6 98  215 314
# *   +   *   +"""
# IN = StringIO(IN_)
IN = open("d6_in.txt")
with IN as f:
    lines = f.read().split("\n")
    lines = [line.strip(" ") for line in lines]
    lines = [re.split(r" +", line) for line in lines]
    ops = lines[-1]
    mat = [[int(i) for i in line] for line in lines[:-1]]
    mat = np.array(mat).T.tolist()

res = 0
for i in range(len(ops)):
    res += (
        reduce(lambda x, y: x+y, mat[i])
        if ops[i] == "+"
        else reduce(lambda x, y: x*y, mat[i])
    )
res
```

Part 2

```python
# from io import StringIO
# IN_ = """123 328  51 64 
#  45 64  387 23 
#   6 98  215 314
# *   +   *   +"""
# IN = StringIO(IN_)
IN = open("d6_in.txt")
with IN as f:
    lines = f.read().split("\n")
    ops = re.split(" +", lines[-1].strip(" "))
    lines = [list(line) for line in lines[:-1]]
    mat = np.array(lines).T.tolist()
    mat = ["".join(line).strip(" ") for line in mat]
    mat = "".join([line+" " if line else "\n" for line in mat])
    mat = mat.split("\n")
    mat = [line.strip(" ").split(" ") for line in mat]
    mat = [[int(i) for i in line] for line in mat]

res = 0
for i in range(len(ops)):
    res += (
        reduce(lambda x, y: x+y, mat[i])
        if ops[i] == "+"
        else reduce(lambda x, y: x*y, mat[i])
    )
res
```
