```python
IN = open("d2_in.txt")
clicks = []
with IN as f:
    ranges = f.readline().rstrip("\r\n").split(",")
    ranges = [r.split("-") for r in ranges]
    ranges_lst = [(int(r[0]), int(r[1])) for r in ranges]
    ranges = [range(r[0], r[1]+1) for r in ranges_lst]

sum = 0
for r in ranges_lst:
    sum += r[1]-r[0]
sum
```

Part 1

```python
count = 0
for r in ranges:
    for i in r:
        s = str(i)
        l = len(s)
        s1, s2 = s[:l//2], s[l//2:]
        if s1 == s2:
            count += i
count
```

Part 2

```python
def is_invalid(s):
    l = len(s)
    for j in range(1, l//2+1):
        s1 = s[:j]
        if (l//j)*s1 == s:
            return int(s)
    return 0
```

```python
count = 0
for r in ranges:
    for i in r:
        s = str(i)
        n = is_invalid(s)
        count += n
count
```
