```python
from collections import deque
```

```python
# from io import StringIO
# IN_ = """3-5
# 10-14
# 16-20
# 12-18

# 1
# 5
# 8
# 11
# 17
# 32"""
# IN = StringIO(IN_)
IN = open("d5_in.txt")
with IN as f:
    lines = f.read().split("\n\n")
    ranges, ids = lines[0].split('\n'), lines[1].split("\n")
    ids = [int(x) for x in ids]
    ranges = [r.split('-') for r in ranges]
    ranges = [(int(r[0]), int(r[1])+1) for r in ranges]
```

Part 1

```python
count = 0
for i in ids:
    for r in ranges:
        if r[0] <= i <= r[1]:
            count += 1
            break
count
```

Part 2

```python
def union_ordered(r, s):
    r1, r2 = r
    s1, s2 = s
    assert r1 <= s1 or (r1 == s1 and r2 <= s2)
    if r2 < s1:
        return [r, s]
    elif r1 <= s1 <= r2 <= s2:
        return [(r1, s2)]
    elif r1 <= s1 <= s2 <= r2:
        return [r]
    return ValueError("Impossible")

def is_disjoint_ordered(r, s):
    r1, r2 = r
    s1, s2 = s
    assert r1 <= s1 or (r1 == s1 and r2 <= s2)
    return r2 < s1
```

```python
count = 0
ranges_sort = deque(sorted(ranges))
while len(ranges_sort) > 1:
    r = ranges_sort.popleft()
    s = ranges_sort.popleft()
    if is_disjoint_ordered(r, s):
        count += r[1] - r[0]
        ranges_sort.appendleft(s)
    else:
        u = union_ordered(r, s)
        assert len(u) == 1
        ranges_sort.appendleft(u[0])
last = ranges_sort.popleft()
count + last[1] - last[0]
```
