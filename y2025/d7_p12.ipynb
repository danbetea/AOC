Part 1

```python
def init(grid):
    pos = grid[0].index("S")
    grid[1][pos] = "|"
    return grid

def split(grid, idx):
    for i in range(1, len(grid[idx])):
        if grid[idx][i] == "^" and grid[idx-1][i] == "|":
            grid[idx][i-1] = "|"
            grid[idx][i+1] = "|"
            grid[idx+1][i-1] = "|"
            grid[idx+1][i+1] = "|"
        elif grid[idx][i] == "." and grid[idx-1][i] == "|":
            grid[idx][i] = "|"
            grid[idx+1][i] = "|"
    return grid

def count_non_hit(grid):
    count = 0
    for i in range(2, len(grid), 2):
        for j in range(len(grid[i])):
            if grid[i][j] == "^" and grid[i-1][j] == ".":
                count += 1
    return count
```

```python
grid = deepcopy(GRID)
num_total = (np.array(grid) == "^").sum()
grid = init(grid)
for idx in range(2, len(grid), 2):
    grid = split(grid, idx)
num_total - count_non_hit(grid)
```

Part 2

```python
@cache
def count(row, col, grid):
    if (row == len(grid)-2 and grid[row][col] == "^" and grid[row-1][col] == "|"):
        return 2
    elif (row == len(grid)-2 and grid[row][col] == "|" and grid[row-1][col] == "|"):
        return 1
    if grid[row][col] == "^" and grid[row-1][col] == "|":
        return count(row+2, col-1, grid) + count(row+2, col+1, grid)
    elif grid[row][col] == "|":
        return count(row+2, col, grid)
    else:
        raise ValueError(f"Impossible at {row},{col},{grid[row][col]}")
```

```python
grid_t = tuple([tuple(g) for g in grid])
r = 2
c = grid[2].index("^")
count(r, c, grid_t)
```
