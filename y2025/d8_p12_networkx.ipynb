Part 1

```python
pairs = list(combinations(POINTS, 2))
pairs = [
    (p, np.linalg.norm(np.array(p[0])-np.array(p[1])).item())
    for p in pairs
]
pairs = sorted(pairs, key=lambda x: x[1])
```

```python
G = nx.Graph()

for p in POINTS:
    G.add_node(p)
for p in pairs[:1000]:
    p1, p2 = p[0][0], p[0][1]
    G.add_edge(p1, p2)
x, y, z = [len(c) for c in sorted(nx.connected_components(G), key=len, reverse=True)][:3]
x*y*z
```

```python
G = nx.Graph()

for p in POINTS:
    G.add_node(p)
for p in pairs:
    p1, p2 = p[0][0], p[0][1]
    G.add_edge(p1, p2)
    if nx.is_connected(G):
        break
p1[0] * p2[0]
```
