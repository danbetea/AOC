Part 1

```python
val = 0
for l in lines:
    m = max(l[:-1])
    nums = []
    for j in range(len(l)-1):
        if l[j] == m:
            n = max(l[j+1:])
            break
    val += m*10+n
val
```

Part 2

```python
def get_max(arr, num_digits):
    if num_digits == 1:
        return arr.max()
    max_idx = np.argmax(arr[:-num_digits+1])
    return arr[max_idx] * 10**(num_digits-1) + get_max(arr[max_idx+1:], num_digits-1)
```

```python
sum_ = 0
for line in lines:
    sum_ += get_max(np.array(line), 12)
sum_
```
