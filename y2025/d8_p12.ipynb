```python
# from io import StringIO
# IN_ = """162,817,812
# 57,618,57
# 906,360,560
# 592,479,940
# 352,342,300
# 466,668,158
# 542,29,236
# 431,825,988
# 739,650,466
# 52,470,668
# 216,146,977
# 819,987,18
# 117,168,530
# 805,96,715
# 346,949,466
# 970,615,88
# 941,993,340
# 862,61,35
# 984,92,344
# 425,690,689"""
# IN = StringIO(IN_)
IN = open("d8_in.txt")
with IN as f:
    lines = f.read().split("\n")
    POINTS = [line.split(",") for line in lines]
    POINTS = [tuple([int(x) for x in line]) for line in POINTS]
```

Part 1

```python
pairs = list(combinations(POINTS, 2))
pairs = [
    (p, np.linalg.norm(np.array(p[0])-np.array(p[1])).item())
    for p in pairs
]
pairs = sorted(pairs, key=lambda x: x[1])
```

```python
circuits = {p: frozenset([p]) for p in POINTS}

for p in pairs[:1000]:
    p1, p2 = p[0][0], p[0][1]
    union = circuits[p1] | circuits[p2]
    for p in union:
        circuits[p] = union
fin_set = set(circuits[k] for k in circuits)
x, y, z = sorted([len(c) for c in fin_set])[-3:]

x * y * z
```

```python
circuits = {p: frozenset([p]) for p in POINTS}

for p in pairs:
    p1, p2 = p[0][0], p[0][1]
    union = circuits[p1] | circuits[p2]
    if len(union) == len(POINTS):
        break
    for p in union:
        circuits[p] = union

p1[0] * p2[0]
```
