```python
# from io import StringIO

# IN_ = """0:
# ###
# ##.
# ##.

# 1:
# ###
# ##.
# .##

# 2:
# .##
# ###
# ##.

# 3:
# ##.
# ###
# ##.

# 4:
# ###
# #..
# ###

# 5:
# ###
# .#.
# ###

# 4x4: 0 0 0 0 2 0
# 12x5: 1 0 1 0 2 2
# 12x5: 1 0 1 0 3 2"""
# IN = StringIO(IN_)

IN = open("d12_in.txt")

CONSTRAINTS = []
with IN as f:
    lines = f.read().split("\n")
    regions = [line for line in lines if len(line) > 3]
    for line in regions:
        size, _, num_shapes = line.partition(": ")
        x, y = size.split("x")
        size = (int(x), int(y))
        num_shapes = tuple(int(x) for x in num_shapes.split(" "))
        CONSTRAINTS.append((size, num_shapes))
        shapes = [line for line in lines if len(line) <= 3]
        SHAPES = "\n".join(shapes).split("\n\n")
        AREAS = tuple(
            (np.array(list(SHAPES[i])) == "#").sum().item()
            for i in range(len(SHAPES))
        )
```

Part 1

```python
count = 0
for c in CONSTRAINTS:
    shape, nums = c
    area = shape[0] * shape[1]
    area_min = np.sum(np.dot(np.array(nums), np.array(AREAS)))
    if area_min < area:
        count += 1
count
```
