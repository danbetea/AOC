```python
def in_bounds(v: tuple[int, int], dims: tuple[int, int]):
    i, j = v
    idim, jdim = dims
    return 0 <= i <= idim-1 and 0 <= j <= jdim-1

def get_st_pos(map_: np.ndarray):
    for i in range(len(map_)):
        for j in range(len(map_[0])):
            if map_[i, j] == '@':
                return (i, j)
    return (-1, -1)

def next_pos(curr: tuple[int, int], d: str, map_: np.ndarray):
    i, j = curr
    match d:
        case '>':
            np = (i, j+1)
        case '<':
            np = (i, j-1)
        case '^':
            np = (i-1, j)
        case 'v':
            np = (i+1, j)
    return np if in_bounds(np, (len(map_), len(map_[0]))) else (-1, -1)
```

```python
def advance_p1(curr: tuple[int, int], d, map_):
    
    i, j = curr
    np = next_pos(curr, d, map_)
    npi, npj = np
    
    if npi == -1:
        raise ValueError("-1 pos")
    
    if map_[npi, npj] == '.':
        map_[i, j] = '.'
        map_[npi, npj] = '@'
        return np
    
    if map_[npi, npj] == '#':
        return curr
    
    if map_[npi, npj] == 'O':
        dx = 0
        if d == '>':
            while map_[npi, npj + dx] == 'O': dx += 1
            if map_[npi, npj + dx] == '#':
                return curr
            elif map_[npi, npj + dx] == '.':
                map_[i, j] = '.'
                map_[npi, npj] = '@'
                map_[npi, npj + dx] = 'O'
                return np
            else:
                raise ValueError('box')
        elif d == '<':
            while map_[npi, npj - dx] == 'O': dx += 1
            if map_[npi, npj - dx] == '#':
                return curr
            elif map_[npi, npj - dx] == '.':
                map_[i, j] = '.'
                map_[npi, npj] = '@'
                map_[npi, npj - dx] = 'O'
                return np
            else:
                raise ValueError('box')
        elif d == '^':
            while map_[npi - dx, npj] == 'O': dx += 1
            if map_[npi - dx, npj] == '#':
                return curr
            elif map_[npi - dx, npj] == '.':
                map_[i, j] = '.'
                map_[npi, npj] = '@'
                map_[npi - dx, npj] = 'O'
                return np
            else:
                raise ValueError('box')
        elif d == 'v':
            while map_[npi + dx, npj] == 'O': dx += 1
            if map_[npi + dx, npj] == '#':
                return curr
            elif map_[npi + dx, npj] == '.':
                map_[i, j] = '.'
                map_[npi, npj] = '@'
                map_[npi + dx, npj] = 'O'
                return np
            else:
                raise ValueError('box')
    raise ValueError("extra options")
    
```

```python
def widen_map(map_: np.ndarray) -> np.ndarray:
    map_lst = map_.tolist()
    new_map = []
    for i in range(len(map_lst)):
        s = ''
        for c in map_lst[i]:
            match c:
                case '#':
                    s += '##'
                case 'O':
                    s += '[]'
                case '.':
                    s += '..'
                case '@':
                    s += '@.'
        new_map.append(list(s))
    return np.array(new_map)
```

```python
def get_2_pos_vert(coords: tuple[tuple[int, int], tuple[int, int]], d: str):
    pl, pr = coords
    if d == 'v':
        new_pl, new_pr = (pl[0]+1, pl[1]), (pr[0]+1, pr[1])
    elif d == '^':
        new_pl, new_pr = (pl[0]-1, pl[1]), (pr[0]-1, pr[1])
    else:
        raise ValueError('bad direction')
    return (new_pl, new_pr)

def is_moveable(coords: tuple[tuple[int, int], tuple[int, int]], d: str, map_: np.ndarray, mv=set()) -> bool:
    """
    ....#
    [][].
    .[]..
    .[]..    this hole thing is moveable one unit up
    [][].    the function builds (with side effects!!!) and mutation
    .[]..    the set of all moveable boxes (tuples (left_corner, right_corner))
    ..[].    and returns true if moveable, false otherwise
    ...[]
    ...@.
    """
    # coordinates of just above or just below box
    pl, pr = get_2_pos_vert(coords, d)
    if map_[pl[0], pl[1]] == '#' or map_[pr[0], pr[1]] == '#':
        return False
    elif map_[pl[0], pl[1]] == '.' and map_[pr[0], pr[1]] == '.':
        mv.add(coords)
        return True
    elif map_[pl[0], pl[1]] == '[' and map_[pr[0], pr[1]] == ']':
        mv.add(coords)
        new_coords = (pl, pr)
        return is_moveable(new_coords, d, map_, mv)
    elif map_[pl[0], pl[1]] == ']' and map_[pr[0], pr[1]] == '.':
        mv.add(coords)
        new_coords = ((pl[0], pl[1]-1), pl)
        return is_moveable(new_coords, d, map_, mv)
    elif map_[pl[0], pl[1]] == '.' and map_[pr[0], pr[1]] == '[':
        mv.add(coords)
        new_coords = (pr, (pr[0], pr[1]+1))
        return is_moveable(new_coords, d, map_, mv)
    elif map_[pl[0], pl[1]] == ']' and map_[pr[0], pr[1]] == '[':
        mv.add(coords)
        # l, r with respect to viewing from above
        new_coords_l = ((pl[0], pl[1]-1), pl)
        new_coords_r = (pr, (pr[0], pr[1]+1))
        return is_moveable(new_coords_l, d, map_, mv) and is_moveable(new_coords_r, d, map_, mv)
    else:
        raise ValueError("Check for extra cases for box moving to handle...")

def move_boxes_vert(box_set: set[tuple[tuple[int, int], tuple[int, int]]], d: str, map_: np.ndarray):
    rev = True if d == 'v' else False  # opposite order, for moving purposes
    
    levels = sorted({pt[0][0] for pt in box_set}, reverse=rev)
    mv_dct = {k: [] for k in levels}
    for pos_tup in box_set:
        level = pos_tup[0][0]
        mv_dct[level].append(pos_tup)
    
    d_inc = 1 if d == 'v' else -1
    for level in levels:
        curr_l = mv_dct[level]
        for c in curr_l:
            pl, pr = c
            pli, plj = pl
            pri, prj = pr
            map_[pli + d_inc, plj] = map_[pli, plj]
            map_[pri + d_inc, prj] = map_[pri, prj]
            map_[pli, plj] = '.'
            map_[pri, prj] = '.'

def advance_p2(curr: tuple[int, int], d: str, map_: np.ndarray):
    i, j = curr
    np = next_pos(curr, d, map_)
    npi, npj = np
    
    if npi == -1:
        raise ValueError("-1 pos")
    
    if map_[npi, npj] == '.':
        map_[i, j] = '.'
        map_[npi, npj] = '@'
        return np
    
    if map_[npi, npj] == '#':
        return curr
    
    if d in '^v':
        coords = ((npi, npj-1), (npi, npj)) if map_[npi, npj] == ']' else ((npi, npj), (npi, npj+1))
        box_set = set()
        if is_moveable(coords, d, map_, box_set):
            move_boxes_vert(box_set, d, map_)
            map_[npi, npj] = '@'
            map_[i, j] = '.'
            return np
        else:
            return curr
    elif d == '>':
        dx = 0
        while map_[npi, npj+dx] == '[' and map_[npi, npj+dx+1] == ']': 
            dx += 2
        if map_[npi, npj+dx] == '#':
            return curr
        elif map_[npi, npj+dx] == '.':
            map_[i, j] = '.'
            map_[npi, npj] = '@'
            for k in range(1, dx+1):
                map_[npi, npj+k] = '[' if k%2 == 1 else ']' 
            return np
        else:
            raise ValueError('box')
    elif d == '<':
        dx = 0
        while map_[npi, npj-dx] == ']' and map_[npi, npj-dx-1] == '[': 
            dx += 2
        if map_[npi, npj-dx] == '#':
            return curr
        elif map_[npi, npj-dx] == '.':
            map_[i, j] = '.'
            map_[npi, npj] = '@'
            for k in range(1, dx+1):
                map_[npi, npj-k] = ']' if k%2 == 1 else '[' 
            return np
        else:
            raise ValueError('box')
```

```python
def pp_color(map_):
    c_map = [['?' for j in range(map_.shape[1])] for i in range(map_.shape[0])]
    for i in range(len(map_)):
        for j in range(len(map_[0])):
            c = map_[i, j]
            if c == '#':
                c_map[i][j] = DARK_GRAY+c
            elif c == '.':
                c_map[i][j] = BRIGHT_CYAN+c
            elif c == ']' or c == '[':
                c_map[i][j] = BRIGHT_GREEN+c
            elif c == '@':
                c_map[i][j] = BRIGHT_RED+c
    for i in range(len(c_map)):
        print(''.join(c_map[i]))
```
